02-DU

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("GTDAnalysis").getOrCreate()

# Load the dataset
gtd_data = spark.read.csv("testtest.csv", header=True, inferSchema=True)
```

```python
# Display the first few rows of the dataset
gtd_data.show()

# Print the schema of the dataset
gtd_data.printSchema()

# Get the number of rows and columns
print(f"The dataset has {gtd_data.count()} rows and {len(gtd_data.columns)} columns.")
```

```python
import matplotlib.pyplot as plt

# Group by year and count the number of incidents
yearly_counts = gtd_data.groupBy("iyear").count().orderBy("iyear").toPandas()

plt.figure(figsize=(14, 7))
plt.plot(yearly_counts["iyear"], yearly_counts["count"], marker='o')
plt.title("Number of Terrorist Incidents Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Incidents")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
from pyspark.sql.functions import *

# Setting aesthetics for plots
sns.set_style("whitegrid")

# 1. Monthly Distribution of Terrorist Incidents
monthly_distribution = gtd_data.groupBy("imonth").count().orderBy("imonth").toPandas()
plt.figure(figsize=(10, 6))
sns.barplot(x="imonth", y="count", data=monthly_distribution)
plt.title("Monthly Distribution of Terrorist Incidents")
plt.xlabel("Month")
plt.ylabel("Number of Incidents")
plt.xticks(ticks=range(12), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.show()
```

```python
from pyspark.sql.types import FloatType

# Convert 'nkill' and 'nwound' columns to float type
gtd_data = gtd_data.withColumn("nkill", gtd_data["nkill"].cast(FloatType()))
gtd_data = gtd_data.withColumn("nwound", gtd_data["nwound"].cast(FloatType()))

# Drop rows where 'nkill' or 'nwound' are NULL (after type casting)
casualties_data = gtd_data.select("nkill", "nwound").dropna().toPandas()

# Now, plot the data as before
plt.figure(figsize=(10, 6))
sns.kdeplot(casualties_data["nkill"], shade=True, label="Killed")
sns.kdeplot(casualties_data["nwound"], shade=True, label="Wounded")
plt.title("Distribution of Casualties")
plt.xlabel("Number of Casualties")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# 3. Top Target Types
top_targets = gtd_data.groupBy("targtype1_txt").count().orderBy(desc("count")).limit(10).toPandas()
plt.figure(figsize=(12, 7))
sns.barplot(x="count", y="targtype1_txt", data=top_targets, palette="viridis")
plt.title("Top 10 Target Types")
plt.xlabel("Number of Incidents")
plt.ylabel("Target Type")
plt.show()
```

```python
# 4. Top Attack Types
top_attacks = gtd_data.groupBy("attacktype1_txt").count().orderBy(desc("count")).limit(10).toPandas()
plt.figure(figsize=(12, 7))
sns.barplot(x="count", y="attacktype1_txt", data=top_attacks, palette="magma")
plt.title("Top 10 Attack Types")
plt.xlabel("Number of Incidents")
plt.ylabel("Attack Type")
plt.show()
```

```python
# 5. Top Terrorist Groups
top_groups = gtd_data.groupBy("gname").count().orderBy(desc("count")).limit(10).toPandas()
plt.figure(figsize=(12, 7))
sns.barplot(x="count", y="gname", data=top_groups, palette="inferno")
plt.title("Top 10 Active Terrorist Groups")
plt.xlabel("Number of Incidents")
plt.ylabel("Terrorist Group Name")
plt.show()
```

```python
# 6. Incidents by Region
region_data = gtd_data.groupBy("region_txt").count().orderBy("count").toPandas()
plt.figure(figsize=(12, 7))
sns.barplot(x="count", y="region_txt", data=region_data, palette="cividis")
plt.title("Number of Incidents by Region")
plt.xlabel("Number of Incidents")
plt.ylabel("Region")
plt.show()
```

```python
# Check for missing values in the dataset
missing_data = gtd_data.select([(gtd_data[c].isNull().cast("int")).alias(c) for c in gtd_data.columns]).groupBy().sum().toPandas()

# Display columns with missing values and their counts
missing_counts = missing_data.transpose()
missing_counts.columns = ["Missing_Count"]
missing_counts = missing_counts[missing_counts["Missing_Count"] > 0]
print(missing_counts)
```

03-DP

```python
selected_columns = ['iyear', 'imonth', 'iday', 'nkill', 'nwound', 'attacktype1_txt', 'targtype1_txt', 'region_txt', 'gname']
gtd_selected = gtd_data.select(selected_columns)
```

```python
gtd_selected.printSchema()
```

```python
gtd_selected.show()
```

```python
# Replace missing or zero values in month and day with 1
gtd_cleaned = gtd_selected.fillna({'imonth': 1, 'iday': 1})

# Replace missing values in nkill and nwound with 0
gtd_cleaned = gtd_cleaned.fillna({'nkill': 0, 'nwound': 0})
```

```python
from pyspark.sql.functions import col

gtd_constructed = gtd_cleaned.withColumn('casualties', col('nkill') + col('nwound'))
```

```python
gtd_constructed.printSchema()
```

```python
gtd_constructed.show()
```

```python
print(f"The dataset has {gtd_constructed.count()} rows and {len(gtd_constructed.columns)} columns.")
```

```python
from pyspark.sql.functions import lit, rand

# Estimate the number of rows for 1993 based on the average rows for surrounding years
num_rows_1992 = gtd_data.filter(gtd_data.iyear == 1992).count()
num_rows_1994 = gtd_data.filter(gtd_data.iyear == 1994).count()
estimated_rows_1993 = (num_rows_1992 + num_rows_1994) // 2

# Randomly generate data for 1993
mock_1993_data = spark.range(estimated_rows_1993).withColumn("iyear", lit(1993))
mock_1993_data = mock_1993_data.withColumn("imonth", (rand() * 12 + 1).cast("int"))
mock_1993_data = mock_1993_data.withColumn("iday", (rand() * 28 + 1).cast("int"))
mock_1993_data = mock_1993_data.withColumn("nkill", (rand() * 100).cast("int"))
mock_1993_data = mock_1993_data.withColumn("nwound", (rand() * 100).cast("int"))

# Drop the 'id' column and Add other txt columns with null values
mock_1993_data = mock_1993_data.drop('id')
mock_1993_data = mock_1993_data.withColumn('attacktype1_txt', lit(None).cast("string"))
mock_1993_data = mock_1993_data.withColumn('targtype1_txt', lit(None).cast("string"))
mock_1993_data = mock_1993_data.withColumn('region_txt', lit(None).cast("string"))
mock_1993_data = mock_1993_data.withColumn('gname', lit(None).cast("string"))

# Calculate the 'casualties' column by adding 'nkill' and 'nwound'
mock_1993_data = mock_1993_data.withColumn('casualties', col('nkill').cast("double") + col('nwound').cast("double"))

# Combine the generated 1993 data with the main GTD dataset
gtd_data_with_1993 = gtd_constructed.union(mock_1993_data)

gtd_data_with_1993.printSchema()
```

```python
print(f"The dataset has {gtd_data_with_1993.count()} rows and {len(gtd_data_with_1993.columns)} columns.")
```

```python
from pyspark.sql.functions import concat_ws

gtd_reformatted = gtd_data_with_1993.withColumn('date', concat_ws('-', 'iyear', 'imonth', 'iday'))

gtd_reformatted.printSchema()
```

```python
gtd_reformatted.show()
```

04-DT

```python
from pyspark.sql.functions import sum as _sum

# Aggregate data to monthly totals for casualties
gtd_monthly = gtd_reformatted.groupBy("iyear", "imonth").agg(_sum("casualties").alias("monthly_casualties")).orderBy("iyear", "imonth")
```

```python
gtd_monthly.printSchema()
```

```python
gtd_monthly.show()
```

```python
print(f"The dataset has {gtd_monthly.count()} rows and {len(gtd_monthly.columns)} columns.")
```

07-DM

```python
# Determine the split point
split_point = int(gtd_monthly.count() * 0.8)

# Splitting the data
train_data = gtd_monthly.limit(split_point)
test_data = gtd_monthly.subtract(train_data)
```

```python
!pip install statsmodels

import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# Convert Spark DataFrame to Pandas DataFrame
train_data_pd = train_data.toPandas()
test_data_pd = test_data.toPandas()

# Use the new ARIMA model implementation without the 'disp' argument
model = ARIMA(train_data_pd['monthly_casualties'], order=(5,1,0))
model_fit = model.fit()
```

```python
forecast = model_fit.forecast(steps=len(test_data_pd)).values
```

08-INT

```python
import matplotlib.pyplot as plt

# Plotting actual vs. forecasted values
plt.figure(figsize=(14, 7))
plt.plot(test_data_pd['monthly_casualties'], label='Actual')
plt.plot(forecast, color='red', label='Forecast')
plt.title('ARIMA Model Forecast vs. Actuals')
plt.legend(loc='upper left')
plt.show()
```

```python
import numpy as np

def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape = mean_absolute_percentage_error(test_data_pd['monthly_casualties'], forecast)
print(f"Mean Absolute Percentage Error (MAPE): {mape}%")
```
